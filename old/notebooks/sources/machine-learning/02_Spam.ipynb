```python
SPAMBASE_NAMES_URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.names'
SPAMBASE_DATA_URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data'

feature_names = [
    line.decode('utf-8').strip().split(':')[0] 
    for line in urlopen(SPAMBASE_NAMES_URL).readlines()[33:]
]

spam_data = pandas.read_csv(SPAMBASE_DATA_URL, header=None, names=(feature_names + ['spam']))
 
X, Y = spam_data.ix[:, :-1].values, spam_data.ix[:, -1].values
 
spam_data.head()
```

```python
TP = ((Y[test] == 1) & (y_pred == 1)).sum()
TN = ((Y[test] == 0) & (y_pred == 0)).sum()
FP = ((Y[test] == 0) & (y_pred == 1)).sum()
FN = ((Y[test] == 1) & (y_pred == 0)).sum()
# Accuracy = (TP + TN) / (TP + TN + FP + FN) — доля правильных ответов
# Precision = TP / (TP + FP) — точность, доля спама среди писем, классифицированных как спам
# Recall = TP / (TP + FN) — полнота, доля спама, которая была отфильтрована
# F1 = 2TP / (2TP + FP + FN) — среднее гармоническое между точностью и полнотой
print('Метрики текущего дерева решения для подразбиения на две последовательные части:')
print('TP: {}  TN: {}  FP: {}  FN: {}'.format(TP, TN, FP, FN))
print('Accuracy: {}'.format((TP + TN) / (TP + TN + FP + FN)))
print('Precision: {}'.format(TP / (TP + FP)))
print('Recall: {}'.format(TP / (TP + FN)))
print('F1: {}'.format(2*TP / (2*TP + FP + FN)))
print('Видимо эти метрики плохо работают в крайних случаях, когда что-то пошло не так. Тут это пустые множества TP и FN.')
print('Стоит перемешать данные.')
print('Так как в этих метриках важна пропорция, то деления на ноль можно починить например прибавив к знаменателям по единице.')
print('Это не сильно скажется на метрике в общем случае, т.к. каждое множество должно быть достаточно велико, чтобы не заметить единицу.')
print('TP: {}  TN: {}  FP: {}  FN: {}'.format(TP, TN, FP, FN))
print('Accuracy: {}'.format((TP + TN) / (TP + TN + FP + FN + 1)))
print('Precision: {}'.format(TP / (TP + FP + 1)))
print('Recall: {}'.format(TP / (TP + FN + 1)))
print('F1: {}'.format(2*TP / (2*TP + FP + FN + 1)))
```

```python
fig = plt.figure(figsize=(15, 6))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

min_precision=0.9
max_recall=0.0
max_recall_info = 'no result'

def process(clf, X_data, name):
    global max_recall, max_recall_info
    results = clf.predict_proba(X_data)
    fpr, tpr, thresholds = roc_curve(Y[test], results[:, 1])
    auc_area = auc(fpr, tpr)
    ax1.plot(fpr, tpr, label=name+' (AUC-ROC={:.2f})'.format(auc_area))
    
    precisions = np.zeros_like(thresholds)
    recalls = np.zeros_like(thresholds)
    for i, threshold in enumerate(thresholds):
        y_pred = (results[:, 1] >= threshold)
        TP = ((Y[test] == 1) & (y_pred == 1)).sum()
        FP = ((Y[test] == 0) & (y_pred == 1)).sum()
        FN = ((Y[test] == 1) & (y_pred == 0)).sum()
        # Precision = TP / (TP + FP + 1) — точность, доля спама среди писем, классифицированных как спам
        # Recall = TP / (TP + FN + 1) — полнота, доля спама, которая была отфильтрована
        precisions[i] = TP / (TP + FP + 1)
        recalls[i] = TP / (TP + FN + 1)
    precisions[0] = 1
    recalls[0] = 0
    
    candidate = recalls[precisions >= min_precision].max()
    if candidate >= max_recall:
        max_recall = candidate
        threshold = thresholds[recalls[precisions >= min_precision].argmax()]
        max_recall_info = name + ' with threshold={:.2f}'.format(threshold)
    ax2.plot(recalls, precisions, label=name)

process(scaled_knn_clf, X_scaled[test], 'scaled KNN')
process(not_scaled_knn_clf, X[test], 'not scaled KNN')
process(scaled_tree_clf, X_scaled[test], 'scaled decision tree')
process(tree_clf, X[test], 'not scaled decision tree')

ax1.plot([0, 1], [0, 1], 'k--', label='random')
ax1.set_xlim([0.0, 1.0])
ax1.set_ylim([0.0, 1.0])
ax1.set_xlabel('False Positive Rate')
ax1.set_ylabel('True Positive Rate')
ax1.legend(loc='lower right')

ax2.set_xlim([0.0, 1.0])
ax2.set_ylim([0.0, 1.0])
ax2.set_xlabel('Recall')
ax2.set_ylabel('Precision')

ax2.plot([0, 1], [min_precision]*2, 'k--', label='precision={}%'.format(100*min_precision))
ax2.plot([max_recall], [min_precision], 'rp')
ax2.legend(loc='lower right')

plt.show()
print('Посмотрим на графики, и подумаем:')
print('Какой максимальной полноты определения спама достигают построенные классификаторы, '\
      'если перед вами стоит жесткое ограничение: точность классификатора не может быть меньше 90%?')
print('Проведем горизонтальную прямую precision=90%, выберем самую правую точку на ней:'\
      ' ее абсциса и является максимальной полнотой: {:.2f}% ({})'.format(100*max_recall, max_recall_info))
```

```python
indicies = np.arange(len(train))
cv_train_part = 0.7 * len(train)
cv_train = train[indicies[:cv_train_part]]
cv_validation = train[indicies[cv_train_part:]]
```

```python
def calculate_auc_roc(clf, X, Y):
    results = clf.predict_proba(X)
    fpr, tpr, thresholds = roc_curve(Y, results[:, 1])
    auc_area = auc(fpr, tpr)
    return auc_area

def create_tree(criterion, max_depth, min_samples_split, min_samples_leaf, X, Y):
    return DecisionTreeClassifier(criterion=criterion, max_depth=max_depth, min_samples_split=min_samples_split,
                                  min_samples_leaf=min_samples_leaf).fit(X, Y)

best_criterion, best_max_depth, best_min_samples_split, best_min_samples_leaf = None, None, None, None
best_auc_roc = 0
    
for criterion in ['gini', 'entropy']:
    for max_depth in range(5, 10):
        for min_samples_split in range(2, 9):
            for min_samples_leaf in range(1, 10):
                clf = create_tree(criterion, max_depth, min_samples_split, min_samples_leaf, X[cv_train], Y[cv_train])
                auc_roc = calculate_auc_roc(clf, X[cv_validation], Y[cv_validation])
                is_better = False
                if auc_roc >= best_auc_roc:
                    is_better = True
                    best_auc_roc = auc_roc
                if is_better:
                    print('criterion={} max_depth={} min_split={} min_leaf={} AUC_ROC={}{}'
                          .format(criterion, max_depth, min_samples_split, min_samples_leaf, auc_roc, ' is better' if is_better else ''))
                    print('AUC_ROC on test: {}'.format(calculate_auc_roc(clf, X[test], Y[test])))
                    best_criterion, best_max_depth, best_min_samples_split, best_min_samples_leaf = criterion, max_depth, min_samples_split, min_samples_leaf
                    
print('Best tree params: criterion={} max_depth={} min_split={} min_leaf={} with AUC_ROC={}'
      .format(best_criterion, best_max_depth, best_min_samples_split, best_min_samples_leaf, best_auc_roc))

best_tree = create_tree(best_criterion, best_max_depth, best_min_samples_split, best_min_samples_leaf, X[train], Y[train])
```

```python
# число соседей, функция расстояния, схема взвешивания

def create_knn(n_neighbors, metric, weights, X, Y):
    return KNN(n_neighbors=n_neighbors, metric=metric, weights=weights).fit(X, Y)

best_n_neighbors, best_metric, best_weights = None, None, None
best_auc_roc = 0
    
for metric in ['euclidean', 'manhattan', 'chebyshev']:
    for weights in ['uniform', 'distance']:
        for n_neighbors in range(1, 20):
            clf = create_knn(n_neighbors, metric, weights, X_scaled[cv_train], Y[cv_train])
            auc_roc = calculate_auc_roc(clf, X_scaled[cv_validation], Y[cv_validation])
            is_better = False
            if auc_roc >= best_auc_roc:
                is_better = True
                best_auc_roc = auc_roc
            if is_better:
                best_n_neighbors, best_metric, best_weights = n_neighbors, metric, weights
                print('metric={} weights={} n_neighbors={} AUC_ROC={} is better'
                      .format(metric, weights, n_neighbors, auc_roc))
                print('AUC_ROC on test: {}'.format(calculate_auc_roc(clf, X_scaled[test], Y[test])))

print('Best KNN params: metric={} weights={} n_neighbors={} with AUC_ROC={}'
      .format(best_metric, best_weights, best_n_neighbors, best_auc_roc))

best_knn = create_knn(best_n_neighbors, best_metric, best_weights, X_scaled[train], Y[train])
```

```python
fig = plt.figure(figsize=(15, 6))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

min_precision=0.9
max_recall=0.0
max_recall_info = 'no result'

def process(clf, X_data, name):
    global max_recall, max_recall_info
    results = clf.predict_proba(X_data)
    fpr, tpr, thresholds = roc_curve(Y[test], results[:, 1])
    auc_area = auc(fpr, tpr)
    ax1.plot(fpr, tpr, label=name+' (AUC-ROC={:.2f})'.format(auc_area))
    
    precisions = np.zeros_like(thresholds)
    recalls = np.zeros_like(thresholds)
    for i, threshold in enumerate(thresholds):
        y_pred = (results[:, 1] >= threshold)
        TP = ((Y[test] == 1) & (y_pred == 1)).sum()
        FP = ((Y[test] == 0) & (y_pred == 1)).sum()
        FN = ((Y[test] == 1) & (y_pred == 0)).sum()
        # Precision = TP / (TP + FP + 1) — точность, доля спама среди писем, классифицированных как спам
        # Recall = TP / (TP + FN + 1) — полнота, доля спама, которая была отфильтрована
        precisions[i] = TP / (TP + FP + 1)
        recalls[i] = TP / (TP + FN + 1)
    precisions[0] = 1
    recalls[0] = 0
    
    candidate = recalls[precisions >= min_precision].max()
    if candidate >= max_recall:
        max_recall = candidate
        threshold = thresholds[recalls[precisions >= min_precision].argmax()]
        max_recall_info = name + ' with threshold={:.2f}'.format(threshold)
    ax2.plot(recalls, precisions, label=name)

process(scaled_knn_clf, X_scaled[test], 'scaled KNN')
process(not_scaled_knn_clf, X[test], 'not scaled KNN')
process(scaled_tree_clf, X_scaled[test], 'scaled decision tree')
process(tree_clf, X[test], 'not scaled decision tree')
process(best_tree, X[test], 'best tree')
process(best_knn, X_scaled[test], 'best knn')

ax1.plot([0, 1], [0, 1], 'k--', label='random')
ax1.set_xlim([0.0, 1.0])
ax1.set_ylim([0.0, 1.0])
ax1.set_xlabel('False Positive Rate')
ax1.set_ylabel('True Positive Rate')
ax1.legend(loc='lower right')

ax2.set_xlim([0.0, 1.0])
ax2.set_ylim([0.0, 1.0])
ax2.set_xlabel('Recall')
ax2.set_ylabel('Precision')

ax2.plot([0, 1], [min_precision]*2, 'k--', label='precision={}%'.format(100*min_precision))
ax2.plot([max_recall], [min_precision], 'rp')
ax2.legend(loc='lower right')

plt.show()
print('Посмотрим на графики, и подумаем:')
print('Какой максимальной полноты определения спама достигают построенные классификаторы, '\
      'если перед вами стоит жесткое ограничение: точность классификатора не может быть меньше 90%?')
print('Проведем горизонтальную прямую precision=90%, выберем самую правую точку на ней:'\
      ' ее абсциса и является максимальной полнотой: {:.2f}% ({})'.format(100*max_recall, max_recall_info))
```
