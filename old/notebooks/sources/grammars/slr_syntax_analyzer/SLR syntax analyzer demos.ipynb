```python
def grammar_tree_image(grammar):
    text = ''
    for i, item in enumerate(grammar.items):
        text += '{} [label="id={}\n{}"];\n'.format(i+1, i, _to_string(item, dot_inlined=True))
    for (i, symbol), j in grammar.goto.items():
        text += '{} -> {} [label="{}"];\n'.format(i+1, j+1, symbol)
    return Image(DotManager.run_dot('digraph mytree {' + text + '}'))
```

```python
def print_syntax_analyze_table(M):
    rows, cols = set(), set()
    for row, col in M.keys():
        rows.add(row)
        cols.add(col)

    print('Syntax analyze table:')
    print('s - shift\nr - reduce\n')

    for col in sorted(cols):
        print('\t{}'.format(col), end='')
    print()
    for row in sorted(rows):
        print('{}\t'.format(row), end='')
        for col in sorted(cols):
            if (row, col) in M:
                action, param = M[row, col]
                if action == 'shift':
                    value = 's{}'.format(param)
                else:
                    if param is None:
                        value = 'r {}'.format('???')
                    else:
                        value = 'r {}→...'.format(param[0])
            else:
                value = ''
            print('{}\t'.format(value), end='')
        print()
```

```python
M = grammar.get_syntax_analyze_table()
print_syntax_analyze_table(M)
```

```python
def parse_tree(grammar, M, tokenizer, debug=False):
    stack = [0]
    symbols = []
    trees = []
    symbol, token = next(tokenizer)
    while True:
        if debug:
            print('stack={}, symbols={}, symbol={}:'.format(stack, symbols, symbol))
        action, param = M[stack[-1], symbol]
        if action == 'shift':
            if debug:
                print('Shift')
            value = None
            if symbol != token:
                symbols.append((symbol, token))
                value = token
            else:
                symbols.append(symbol)
            trees.append(Tree(symbol, value=value))
            stack.append(param)
            symbol, token = next(tokenizer)
        elif action == 'reduce':
            if debug:
                print('Reduce')
            key, rule, foo = param
            symbols = symbols[:-len(rule)] + [key]
            trees = trees[:-len(rule)] + [Tree(key, child_trees=trees[-len(rule):], foo=foo)]
            stack = stack[:-len(rule)]
            stack.append(grammar.goto[stack[-1], key])
        else:
            assert action == 'accept'
            if debug:
                print('Accepted')
            break
    tree = trees[0]
    for i in range(239):
        tree.calculate()
    return tree
```

```python
M = grammar.get_syntax_analyze_table()
lexical_analyzer = ExpressionLexicalAnalyzer()

string = 'a*b+c*x'
tokenizer = lexical_analyzer.parse(string)
tree = parse_tree(grammar, M, tokenizer)

Image(DotManager.run_dot('digraph mytree {' + tree.to_text() + '}'))
```

# 3. Latex grammar (with convertation to MathML)

```python
grammar3 = Grammar(latex_grammar.start, latex_grammar.rules, latex_grammar.terminals)
grammar_tree_image(grammar3)
```

```python
string = '\sqrt{a^2+b_1-x^\sqrt{y_5+717/3}+2*39}'
tokenizer = LatexLexicalAnalyzer().parse(string)
tree3 = parse_tree(grammar3, grammar3.get_syntax_analyze_table(), tokenizer)

Image(DotManager.run_dot('digraph mytree {' + tree3.to_text(with_empty=True) + '}'))
```

```python
def to_math_ml(tree, tab=0, root=True):
    offset_str = '  '
    text = ''
    if root:
        text += offset_str * tab + '<math display="block">\n'
        tab += 1
        
    tag = None
    value = None
    before_child = {}
    to_increment = 1
    if tree.key == 'sqrtExpr':
        tag = 'msqrt'
    elif tree.key == 'indexedExpr' and len(tree.child_trees) == 3:
        if tree.child_trees[1].key == '_':
            tag = 'msub'
        else:
            assert tree.child_trees[1].key == '^'
            tag = 'msup'
    elif tree.key == 'Variable':
        tag = 'mi'
        value = tree.value
    elif tree.key == 'Number':
        tag = 'mn'
        value = tree.value
    elif tree.key in {'addExpr', 'multExpr'} and len(tree.child_trees) == 3:
        between = '<mo>{}</mo>\n'.format(tree.child_trees[1].key)
        before_child[1] = between
    else:
        to_increment = 0
    
    if tag:
        text += offset_str * tab + '<{}>\n'.format(tag)
        
    if value:
        text += offset_str * (tab + to_increment) + '{}\n'.format(value)
        
    for i, child in enumerate(tree.child_trees):
        if i in before_child:
            text += offset_str * (tab + to_increment) + before_child[i]
        text += to_math_ml(child, tab + to_increment, root=False)
        
    if tag:
        text += offset_str * tab + '</{}>\n'.format(tag)
    if root:
        tab -= 1
        text += offset_str * tab + '</math>\n'
    return text
```

Note, that MathML currently supported only by Mozilla Firefox.

To ensure that your browser supports MathML: https://developer.mozilla.org/en-US/docs/Mozilla/MathML_Project/MathML_Torture_Test

```python
HTML(to_math_ml(tree3))
```

This is how TEX renders this formula:
$$\sqrt{a^2 + b_1-x^\sqrt{y_5+717/3} + 2*39}$$
