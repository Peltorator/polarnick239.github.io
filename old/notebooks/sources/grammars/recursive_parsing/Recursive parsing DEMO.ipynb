```python
from recursive_parsing.grammars import pascal_vars_grammar
from recursive_parsing.grammars.grammar import Grammar
from recursive_parsing.lexical_analyzers.pascal_lexical_analyzer import PascalLexicalAnalyzer
from recursive_parsing.syntax_analyzer import syntax_analyze

from IPython.display import Image
from ctree.visual.dot_manager import DotManager
```

```python
start = pascal_vars_grammar.start
rules = pascal_vars_grammar.rules
terminals = pascal_vars_grammar.terminals
grammar = Grammar(start, rules, terminals)
```

```python
print('Starting non-terminal:\n{}\n'.format(start))

print('Raw rules:\n')

for key, key_rules in rules.items():
    for rule in key_rules:
        print(' {} -> {}'.format(key, rule))

print('\nRules:\n')
for key, key_rules in grammar.rules.items():
    for rule in key_rules:
        print(' {} -> '.format(key), end='')
        for token in rule:
            print('{} '.format(token), end='')
        print()
```

```python
print('First:')
for key, first in grammar.first.items():
    if key in grammar.terminals:
        continue
    tokens = list(first)
    string = "'{}'".format(tokens[0])
    for token in tokens[1:]:
        string += ", '{}'".format(token)
    print(" {}: {}".format(key, string))
```

```python
print('Follow:')
for key, follow in grammar.follow.items():
    if key in grammar.terminals:
        continue
    tokens = list(follow)
    string = "'{}'".format(tokens[0])
    for token in tokens[1:]:
        string += ", '{}'".format(token)
    print(" {}: {}".format(key, string))
```

```python
small_string = 'var a, b: integer; c: integer;'

small_tree = syntax_analyze(small_string, PascalLexicalAnalyzer(), grammar)
text = 'digraph mytree {' + small_tree.to_text(with_empty=False) + '}'

Image(DotManager.run_dot(text))
```

```python
def iter_filtered(tree, is_intresting, is_go_on=None):
    if is_intresting(tree):
        yield tree
    for child in tree.child_trees:
        if is_go_on is None or is_go_on(child):
            yield from iter_filtered(child, is_intresting)
```

```python
string = '''
var
    ai, bi, ci : integer;
    ar, br, cr : real;
    ac, bc, cc, dc : char;
    ab, bb : boolean;
    extrai : integer;
    extrareal : real;
'''
print(string)

tree = syntax_analyze(string, PascalLexicalAnalyzer(), grammar)

nodes_text = ''
connections_text = ''
next_id = 1

merge = True
type_to_id = {}

for node in iter_filtered(tree, lambda tree: tree.key == 'type'):
    if merge:
        if node.value in type_to_id:
            type_id = type_to_id[node.value]
        else:
            type_id = next_id
            next_id += 1
            nodes_text += '{} [label="{}"];'.format(type_id, node.value)
            type_to_id[node.value] = type_id
    else:
        type_id = next_id
        next_id += 1
        nodes_text += '{} [label="{}"];'.format(type_id, node.value)
    for var in iter_filtered(node.parent, lambda tree: tree.key == 'name', lambda tree: tree.key != 'Variables'):
        connections_text
        nodes_text += '{} [label="{}"];'.format(next_id, var.value)
        connections_text += '{} -> {};'.format(type_id, next_id)
        next_id += 1

text_tree = 'digraph mytree {' + nodes_text + connections_text + '}'

Image(DotManager.run_dot(text_tree))
```
