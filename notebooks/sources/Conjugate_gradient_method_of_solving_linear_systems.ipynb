```python
# A = array([
# [2, 1, -100],
# [1, -5, 1000],
# [-100, 1000, 3]
# ])
# b = array([-296, 2991, 1909])
# solution = array([1, 2, 3])

A = array([
[4, 2],
[2, 2]
])
b = array([16, 10])
solution = array([3, 2])

assert np.linalg.det(A) != 0
assert solution is None or np.all(A.dot(solution) == b)
assert np.all(A == A.T)

print('Linear system to solve:\n')
for i in range(A.shape[0]):
    print('{}*x0'.format(A[i, 0]), end='')
    for j in range(A.shape[1]-1):
        print(' + {}*x{}'.format(A[i, j+1], j+1), end='')
    print(' = {}'.format(b[i]))

if solution is not None:
    print('\nSystem solution:\n')
    print('x0 = {}'.format(solution[0]), end='')
    for i in range(len(solution) - 1):
        print(', x{} = {}'.format(i+1, solution[i+1]), end='')
```

```python
x_old = np.zeros(b.shape[0])
r_old = b - A.dot(x_old)
z_old = r_old

steps = A.shape[0]
draw_isolines = (A.shape[0] == 2)

if draw_isolines:
    count = 100
    xs, ys = np.mgrid[0:count, 0:count] / count
    if solution is not None:
        xs, ys = xs.ravel() * solution[0] * 2, ys.ravel() * solution[1] * 2
    else:
        xs, ys = xs.ravel() * solution[0] * 100 * 2 - 50, ys.ravel() * 100 * 2 - 50
    plt.contour(xs.reshape(count, count), ys.reshape(count, count),
                ((b - A.dot([xs, ys]).T) * (b - A.dot([xs, ys]).T)).sum(axis=1).reshape(count, count),
                levels=np.power(np.arange(0.1, 50.0, 0.4), 2.0));

for i in range(3 * steps):
    print('Step {}...'.format(i + 1))
    alpha = (r_old * r_old).sum() / (A.dot(z_old) * z_old).sum()
    x = x_old + alpha * z_old
    r = r_old - alpha * A.dot(z_old)
    if (np.sum(r * r) / np.sum(b * b) < 1e-32):
        break
    beta = (r * r).sum() / (r_old * r_old).sum()
    z = r + beta * z_old
    if draw_isolines:
        plt.plot([x_old[0], x[0]], [x_old[1], x[1]], color='red', linestyle='-', linewidth=1)
    x_old, r_old, z_old = x, r, z
    
print('\nFounded solution:\n')
print('x0 = {:.6f}'.format(x[0]), end='')
for i in range(len(x) - 1):
    print(', x{} = {:.6f}'.format(i+1, x[i+1]), end='')

print('\n\nSum of squared errors: {}'.format(((b - A.dot(x)) * (b - A.dot(x))).sum()))
```
