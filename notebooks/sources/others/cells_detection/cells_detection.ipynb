```python
import cv2

import numpy as np

import matplotlib as mpl
from matplotlib import pyplot as plt

import math
import functools

%matplotlib inline

# author: Polyarniy Nickolay
```

```python
plt.rcParams['figure.figsize'] = 20, 30
imshow = plt.imshow
imshow_gray = functools.partial(plt.imshow, cmap=mpl.cm.Greys_r)
```

```python
filename = 'second_day_p01t003_c1.jpg'
img_full = cv2.imread(filename)
```

```python
imshow(img_full);
```

```python
img_left_part = np.rot90(img_full[:, :500])
imshow(img_left_part);
```

```python
img_canny_thresholded = cv2.Canny(img_left_part[:, :, 0], 200, 250)
imshow_gray(img_canny_thresholded);
```

```python
img_gaussian_thresholded = cv2.adaptiveThreshold(img_left_part[:, :, 0], 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 9, 17)
imshow_gray(img_gaussian_thresholded);
```

```python
kernel_size = 2
img_opened_gaussian = cv2.morphologyEx(img_gaussian_thresholded, cv2.MORPH_OPEN, np.ones((kernel_size, kernel_size),np.uint8))
img_opened_gaussian = cv2.morphologyEx(img_opened_gaussian, cv2.MORPH_DILATE, np.ones((kernel_size*2, kernel_size*2),np.uint8))
imshow_gray(img_opened_gaussian);
```

```python
tmp, contours, hierarchy = cv2.findContours(img_gaussian_thresholded, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
img = np.ascontiguousarray(img_left_part)
for i in range(len(contours)):
    img = cv2.drawContours(img, contours, i, (0, 255, 0))
imshow(img);
```

```python
img = np.ascontiguousarray(np.array(img_left_part, np.float64))
ellipses = []
for cnt in contours:
    if len(cnt) < 10:
        continue
    rount_to_int = lambda x: int(math.round(x))
    contour_area = cv2.contourArea(cnt)
    contour_perimeter = cv2.arcLength(cnt, closed=True)
    
    ellipse = cv2.fitEllipse(cnt)
    center = tuple(map(int, ellipse[0]))
    major_minor_axis = tuple(map(int, ellipse[1]))
    angle = int(ellipse[2])
    ellipse_bb_area = major_minor_axis[0] * major_minor_axis[1]
    ellipse_bb_perimeter = (major_minor_axis[0] + major_minor_axis[1]) * 2.0
    area_ratio = ellipse_bb_area / contour_area
    perimeter_ratio = ellipse_bb_area / contour_perimeter
    if max(major_minor_axis) > 4.0 * min(major_minor_axis):
        continue
    if area_ratio > 2.0 or area_ratio < 0.5:
        continue
    ellipses.append((center, major_minor_axis, angle))
    img = cv2.ellipse(img, center, major_minor_axis, angle, 0, 360, (0, 0, 125), thickness=2)
imshow(img);
```

```python
merged_ellipses = []
for e1 in ellipses:
    c1, mma1, a1 = e1
    matched = False
    for e2 in merged_ellipses:
        c2, ma2, a2 = e2
        if np.all(np.abs(np.array(c2) - np.array(c1)) < 5.0):
            matched = True
    if not matched:
        merged_ellipses.append(e1)
print('{} merged to {}'.format(len(ellipses), len(merged_ellipses)))
```

```python
img = np.ascontiguousarray(np.array(img_left_part, np.float64))
for e in merged_ellipses:
    img = cv2.ellipse(img, e[0], e[1], e[2], 0, 360, (0, 0, 125), thickness=2)
imshow(img);
```
