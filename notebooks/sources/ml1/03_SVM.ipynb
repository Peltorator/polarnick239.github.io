```python
x1 = np.linspace(-3, 5, 50)
x2 = np.linspace(-2, 6, 50)
xx1, xx2 = np.meshgrid(x1, x2)

f = -8*xx1 - 16*xx2 + xx1**2 + 4*xx2**2
constr1 = (xx1 + xx2 <= 5)
constr2 = (0 <= xx1) & (xx1 <= 3)
constr3 = (xx2 >= 0)

plt.figure(figsize=(5, 5))
plt.grid(True)
plt.xlabel('x1')
plt.ylabel('x2')

plt.contour(x1, x2, f, 50, alpha=0.5)
plt.contourf(x1, x2, constr1 & constr2 & constr3, 1, alpha=0.5, cmap='gray')
_ = plt.scatter([3], [2])
```

```python
P = cvx.matrix([[2., 0.],
            [0., 8.]])
q = cvx.matrix([-8., -16.])

G = cvx.matrix([[1., 1.],     # x1 + x2 <= 5
            [0., -1.],    # -x2 <= 0
            [1., 0.],     #  x1 <= 3
            [-1., 0.]]).T # -x1 <= 0
h = cvx.matrix([5., 0., 3., 0.])

print('Минимизируем функцию f(x)=−8*x1 − 16*x2 + x1*x1 + 4*x2*x2')
print('С ограничениями: x1 + x2 ≤ 5, 0 ≤ x1 ≤ 3, x2 ≥ 0')
solution = cvx.solvers.qp(P, q, G, h)
xs = solution['x']

print('Найденное решение: x1={}, x2={}. Совпадает с решением, видимым из рисунка.'.format(xs[0], xs[1]))
```

```python
print('Из рисунка видно, что точку оптимума от движения вправо к глобальному минимуму останавливает только правая грань.')
print('Т.е. активным является только x1 <= 3.')
print('На всякий случай проверим:')
P = cvx.matrix([[2., 0.],
            [0., 8.]])
q = cvx.matrix([-8., -16.])

G = cvx.matrix([[1., 0.]]).T #  x1 <= 3
h = cvx.matrix([3.])

solution = cvx.solvers.qp(P, q, G, h)
xs = solution['x']

print('Найденное решение: x1={}, x2={}. Действительно, оставив только ограничение в x1 <= 3, точка оптимума не изменилась.'.format(xs[0], xs[1]))
```

```python
# Реализуем свой BaseEstimator, который будет формулировать задачу линейного SVM для QP солвера.
# Матрицы будет делать разреженными, т.к. это дает прирост скорости в 10 раз.

class LinearSVM(BaseEstimator):
    def __init__(self, C, transform=None):
        self.C = C
        if transform is None:
            transform = lambda x: x
        self.transform = transform
        
    def fit(self, X, y):
        X = self.transform(X)
        n, dim = X.shape
        
        self.bias_ = 0 # x0
        self.weights_ = np.zeros(dim) # x1, x2, x3, ... x(dim-1), xdim
        self.slack_variables = np.zeros(n) # x(dim+1), x(dim+2), ..., x(dim+n-1), x(dim+n)
        
        P = np.diag([0] + [1.0] * dim + [0] * n)
        q = np.array([0] * (1 + dim) + [self.C] * n)
        
        # slack_variables >= 0
        G1 = np.hstack((np.zeros((n, (1 + dim))), np.diag([-1.0]*n)))
        h1 = np.array([0.0]*n)
        
        # yn(wT*xn+b)≥1−ξn
        # -yn(wT*xn+b) <= ξn-1
        # -yn(wT*xn+b)-ξn <= -1
        # -(yn(wT*xn+b)+ξn) <= -1
        G2list = []
        for i in range(0, n):
            # yn(wT*xn+b)+ξn
            yn_b = np.array([y[i]])
            yn_w_xn = y[i] * X[i]
            slack_vars = np.zeros(n)
            slack_vars[i] = 1.0
            G2list.append(np.hstack((yn_b, yn_w_xn, slack_vars)))
        G2 = -np.array(G2list)
        h2 = np.array([-1.0] * n)
        
        G = np.vstack((G1, G2))
        h = np.hstack((h1, h2))
        
        # Использование разреженных матриц дает прирост больше чем в десять раз (66 мс. против 720 мс.)
        solution = cvx.solvers.qp(cvx.sparse(cvx.matrix(P)), cvx.matrix(q), cvx.sparse(cvx.matrix(G)), cvx.matrix(h))
        xs = np.array(solution['x']).T[0]
        
        self.bias_ = xs[0]
        self.weights_ = xs[1:(dim+1)]
        self.slack_variables = xs[(dim+1):(dim+n+1)]

        # Построим маску для визуализации опорных объектов
        y_predict = self.predict_proba(X)
        tmp = np.sum(self.weights_*X, axis=1)
        vals = y_predict * (tmp + self.bias_) + self.slack_variables - 1.0
        EPSILON = 0.000001
        self.support_mask = vals <= EPSILON
        
    def predict_proba(self, X):
        X = self.transform(X)
        y_predict = (np.dot(X, self.weights_) + self.bias_).flatten()
        return y_predict
            
    def predict(self, X):
        result = np.sign(self.predict_proba(X))
        result[result == 0] = 1.0
        return result
```

```python
# Напишем функцию, которая будет визуализировать вероятность класса,
# которую оценивает переданный LinearSVM на переданных 2D данных

def visualize(clf, X, y, quality=400):
    X0 = X[:, 0]
    X1 = X[:, 1]
    
    border = .2
    X0range = X0.max() - X0.min()
    X1range = X1.max() - X1.min()
    X0from, X0to = X0.min() - X0range * border, X0.max() + X0range * border
    X1from, X1to = X1.min() - X1range * border, X1.max() + X1range * border
    x0vals = np.linspace(X0from, X0to, quality)
    x1vals = np.linspace(X1from, X1to, quality)

    xs0, xs1 = np.meshgrid(x0vals, x1vals)
    ys = np.reshape(clf.predict_proba(np.vstack((xs0.ravel(), xs1.ravel())).T), xs0.shape)

    # Цветной фон, соответствующий значению решающего правила
    plt.pcolormesh(xs0, xs1, ys, cmap='autumn', zorder=0)

    # Линии уровня
    matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
    plt.contour(xs0, xs1, ys, levels=(0.0,), colors='k', linewidths=1.5, zorder=1)
    plt.contour(xs0, xs1, ys, levels=(-1.0,), colors='k', linewidths=0.5, zorder=1)
    plt.contour(xs0, xs1, ys, levels=(1.0,), colors='k', linewidths=0.5, zorder=1)

    # yn(wT⋆xn+b)=1−ξn    
    y_predict = clf.predict(X)
    support_mask = clf.support_mask.copy()
    wrong_mask = y_predict != y
    support_mask[wrong_mask] = False
    others_mask = np.ones_like(support_mask, np.bool)
    others_mask[support_mask] = False
    others_mask[wrong_mask] = False
    
    def draw_points(mask, marker, size):
        pos_mask = y > 0
        mask1 = np.logical_and(mask, pos_mask)
        plt.scatter(X0[mask1], X1[mask1], c='#ffff00', marker=marker, s=size)
        mask2 = np.logical_and(mask, ~pos_mask)
        plt.scatter(X0[mask2], X1[mask2], c='r', marker=marker, s=size)
    
    draw_points(others_mask, 's', 10)
    draw_points(support_mask, 'o', 20)
    draw_points(wrong_mask, 'x', 20)
    
    plt.xlim([X0from, X0to])
    plt.ylim([X1from, X1to])
```

```python
# Выключим вывод результатов каждого шага итераций в stdout QP solver-а
cvx.solvers.options['show_progress'] = False
```

```python
X1, y1 = make_classification(n_samples=100, n_features=2, n_redundant=0, n_clusters_per_class=1, random_state=1)
X1 += 100.0
y1[y1==0] = -1

def check_simple(C10pow):
    C = math.exp(C10pow)
    print('Сгенерируем хорошо разделимую выборку.')
    clf = LinearSVM(C)
    clf.fit(X1, y1)

    visualize(clf, X1, y1)
    print('Видно, что SVM работает верно. (Кружочки - опорные)')
    
interact(check_simple, C10pow=(-3, 3, 1));
```

```python
X2, y2 = make_classification(n_samples=2000, n_features=2, n_redundant=0, n_clusters_per_class=2, random_state=3)
X2 += 100.0
y2[y2==0] = -1

def check_bad(C10pow):
    C = 10. ** C10pow
    print('Сгенерируем объекты с плохо разделимыми классами по имеющимся признакам.')
    converter = lambda x: np.vstack([x[:, 0], x[:, 1], x[:, 0]*x[:, 1]]).T
    clf = LinearSVM(C, converter)
    clf.fit(X2, y2)

    visualize(clf, X2, y2)
    print('Видно, что SVM делает все, что может, но ошибки есть, т.к. выборка не разделима (есть выбросы - элементы другого класса). (Кружочки - опорные, крестики - выбросы, квадраты - остальные)')
    
interact(check_bad, C10pow=(-5, 10, 1));
```

```python
print('Сгенерируем объекты с хорошо разделимыми классами, но не линейно.')

X3, y3 = make_circles(n_samples=1000, noise=0.12, factor=0.5, random_state=1)
X3 += 10.0
y3[y3==0] = -1

def check_circles(C10pow):
    C = 10. ** C10pow
    converter = lambda x: np.vstack([x[:, 0], x[:, 1], x[:, 0]*x[:,0], x[:, 1]*x[:, 1]]).T
    clf = LinearSVM(C, converter)
    clf.fit(X3, y3)

    visualize(clf, X3, y3)
    print('Это пример решения SVM квадратичной поверхностью. (Кружочки - опорные, крестики - выбросы, квадраты - остальные)')
    
interact(check_circles, C10pow=(-2, 10, 1));
```

```python
print('Видно, что чем больше параметр C, тем ширина полосы уже (меньше опорных векторов), и соответственно меньше ошибок, т.к. за них больше штраф')
print('если же параметр C меньше, то SVM становится толерантнее к выбросам, а значит меньше риск переобучения (и больше риск недообучения).')
```

```python
# Реализуем ядровой SVM

def kernel_linear():
    def foo(X1, X2):
        return np.dot(X1, X2)
    return foo

def kernel_polynomial(d=2):
    def foo(X1, X2):
        return (np.dot(X1, X2) + 1) ** d
    return foo

def kernel_gauss_RBF(sigma=1.0):
    def foo(X1, X2):
        norm2 = np.linalg.norm(X1-X2)**2
        return np.exp(-norm2/(2.*sigma))
    return foo

class KernelSVM(BaseEstimator):
    def __init__(self, C, kernel=kernel_linear()):
        self.C = C
        self.kernel = kernel
        
    def fit(self, X, y):
        n, dim = X.shape
        
        self._n, self._dim = n, dim
        self._y = y
        self._X = X
        
        self.bias_ = 0 # x0
        self.weights_ = np.zeros(dim) # x1, x2, x3, ... x(dim-1), xdim
        self.slack_variables = np.zeros(n) # x(dim+1), x(dim+2), ..., x(dim+n-1), x(dim+n)
        
        K = np.zeros((n, n), np.float)
        P = np.zeros((n, n), np.float)
        for i in range(0, n):
            for j in range(0, n):
                K[i, j] = self.kernel(X[i, :], X[j, :])
                P[i, j] = y[i] * y[j] * K[i, j]
        q = np.array([-1.] * n)
        
        G = np.vstack((np.diag([1.]*n), np.diag([-1.]*n)))
        h = np.array([self.C]*n + [0.]*n)
        
        A = np.array([y.T], np.float)
        b = np.array([0.])
        
        # Использование разреженных матриц дает прирост больше чем в десять раз (66 мс. против 720 мс.)
        solution = cvx.solvers.qp(cvx.sparse(cvx.matrix(P)), cvx.matrix(q), cvx.sparse(cvx.matrix(G)), cvx.matrix(h),
                                  cvx.matrix(A), cvx.matrix(b))
        self._xs = np.array(solution['x']).T[0]
        
        EPSILON = 0.000001
        self.support_mask = self._xs > EPSILON
        
        self._biases = []
        for i in range(0, n):
            bias = y[i] - self._xs[i]*y[i]*np.sum(K[:, i])
            self._biases.append(bias)
        self.bias = np.average(self._biases)
        
    def predict_proba(self, X):
        m = X.shape[0]
        proba = np.zeros(m)
        for i in range(0, m):
            for j in range(0, self._n):
                proba[i] += self._xs[j] * self._y[j] * self.kernel(X[i, :], self._X[j, :])
            proba[i] += self.bias
        return proba
            
    def predict(self, X):
        result = np.sign(self.predict_proba(X))
        result[result == 0] = 1.0
        return result
```

```python
X4, y4 = make_classification(n_samples=100, n_features=2, n_redundant=0, n_clusters_per_class=1, random_state=1)
X4 += 100.0
y4[y4==0] = -1

def check_kernel_simple(C10pow):
    print('Сгенерируем хорошо разделимую выборку.')
    C = 10. ** C10pow
    clf = KernelSVM(C, kernel_linear())
    clf.fit(X4, y4)

    visualize(clf, X4, y4, 30)
    print('Видно, что ядровой SVM работает на ней так же верно, но медленнее (пришлось реже сетку делать для цвета фона). (Кружочки - опорные)')
    
interact(check_kernel_simple, C10pow=(-3, 3, 1));
```

```python
print('Сгенерируем объекты с хорошо разделимыми классами, но не линейно.')

X5, y5 = make_circles(n_samples=100, noise=0.12, factor=0.5, random_state=1)
X5 += 100.0
y5[y5==0] = -1

def check_bad(C10pow, d, verbose=False):
    C = 10. ** C10pow
    if verbose:
        print('Сгенерируем объекты с хорошо разделимыми классами, но не линейно. И применим к ним ядровой SVM с полиномиальным ядром.')
    clf = KernelSVM(C, kernel_polynomial(d))
    clf.fit(X5, y5)

    visualize(clf, X5, y5, 30)
    if verbose:
        print('Малая степень полинома приводит к недообучению, что можно компенсировать увеличением C.')
        print('Про C - так же, как было и везде - чем меньше - тем глаже граница и больше коридор (недообучение), чем больше - тем уже коридор и резче граница (переобучение).')

interact(check_bad, C10pow=(-5, 15, 1), d=(2, 3, 1), verbose=True);
```

```python
X6, y6 = make_classification(n_samples=200, n_features=2, n_redundant=0, n_clusters_per_class=2, random_state=5)
X6 += 100.0
y6[y6==0] = -1

def check_bad(C10pow, sigma10pow, verbose=False):
    C = 10. ** C10pow
    sigma = 10. ** sigma10pow
    if verbose:
        print('Сгенерируем объекты с плохо разделимыми классами по имеющимся признакам.')
    clf = KernelSVM(C, kernel_gauss_RBF(sigma))
    clf.fit(X6, y6)

    visualize(clf, X6, y6, 30)
    if verbose:
        print('Видно, что ядровой SVM с Гауссовским-RBF ядром уже может как-то разумно разделить даже такие выборки. (Кружочки - опорные, крестики - выбросы, квадраты - остальные)')
        print('Параметр сигма: чем меньше, тем больше ядро учитывает каждую точку, соответственно тяготеет к переобучению.')
        print(' чем меньше параметр сигма, тем более гладким становится ядро, соответственно тяготеет к недообучению.')
        print('Про C - так же, как было и раньше - чем меньше - тем глаже граница и больше коридор (недообучение), чем больше - тем уже коридор и резче граница (переобучение).')

interact(check_bad, C10pow=(-5, 10, 1), sigma10pow=(-5, 7, 1), verbose=True);
```
