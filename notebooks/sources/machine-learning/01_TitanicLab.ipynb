```python
plt.figure(figsize=(18,6), dpi=1000);

plt.subplot2grid((1, 2), (0, 0))
def plot_survivors_per_class():
    classes = df['Pclass']
    survived = (df['Survived'] == 1)
    unique_classes = np.unique(classes)
    survived_per_class = [np.logical_and(classes == c, survived).sum() for c in unique_classes]
    plt.bar(unique_classes - .4, survived_per_class, label='survived', color='green', alpha=.4)
    not_survived_per_class = [np.logical_and(classes == c, ~survived).sum() for c in unique_classes]
    plt.bar(unique_classes - .4, not_survived_per_class, label='not survived', bottom=survived_per_class, color='red', alpha=0.4)
    plt.xticks(unique_classes, ['{} class'.format(c) for c in unique_classes])
    plt.ylabel('Passangers count')
    plt.legend(loc='upper left')
plot_survivors_per_class()

classes = df['Pclass']
firstClass = (classes == 1).sum()
print('In 1 class were {} passangers. That is {}% of all passangers.'.format(firstClass, int(100 * firstClass / classes.shape[0])))

plt.subplot2grid((1, 2), (0, 1))
ages = df['Age']
plt.hist(ages[ages.notnull()], bins=ages.max() - ages.min() + 1, alpha=0.4)
plt.xlabel('Age')
plt.ylabel('Passangers count');
print('Passangers\' age was distributed from {} to {} years, with median value - {} years old.'.format(ages.min(), int(ages.max()), int(ages.median())))
```

```python
matches = [re.search(r"[\w]+\.", name) for name in df['Name']]
titles = array(list(map(lambda match: None if match is None else match.group(0), matches)))
classes = array(df['Pclass'], np.int)

fig = plt.figure(figsize=(18, 12), dpi=1300)

for clazz in np.sort(np.unique(classes)):
    clazz_titles = itemfreq(titles[classes == clazz])
    clazz_titles = clazz_titles[np.argsort(clazz_titles[:, 1].astype(np.int))]
    
    ax = plt.subplot2grid((len(np.unique(classes)), 1), (clazz - 1, 0))
    plt.bar(np.arange(len(clazz_titles[:, 0])) - .4, clazz_titles[:, 1].astype(np.int), alpha=.4)
    plt.xticks(np.arange(len(clazz_titles[:, 0])), clazz_titles[:, 0])
    plt.xlabel('Titles')
    plt.ylabel('Passengers count')
    plt.title('Passengers in {} class:'.format(clazz));
    for i, (_, count) in enumerate(clazz_titles):
        y = int(clazz_titles[i, 1])
        ax.text(i, y, y)
    ax.set_yscale('log')
fig.tight_layout()
```

```python
matches = [re.findall(r"[\d]+", ticket) for ticket in df['Ticket']]
tickets = array(list(map(lambda match: None if len(match) == 0 else int(match[-1]), matches)))

plt.figure(figsize=(18,6), dpi=1000)
classes = array(df['Pclass'])
filtered = {}
for clazz in np.unique(classes):
    mask = (classes == clazz)
    tickets_no_none = array(list(filter(None, tickets[mask])))
    filtered[clazz] = tickets_no_none
    plt.plot(classes[mask] + np.random.random(classes[mask].shape) * .2, tickets[mask], 'bo',
             color=mpl.colors.hsv_to_rgb((clazz/len(np.unique(classes)), 0.5, 0.5)), label='{} class'.format(clazz), alpha=0.2);
plt.legend(loc='upper left')
plt.ylabel('Ticket number')
plt.xticks(np.unique(classes) + .1, ['{} class'.format(c) for c in np.unique(classes)]);

print('Tickets - class dependencies:\n')

for clazz in [1, 2]:
    print('{} class:'.format(clazz))
    t = filtered[clazz]
    threshold = (t.min() + t.max()) // 2
    low, high = t[t < threshold], t[t >= threshold]
    print(' Two groups:')
    print('  {}.1: around {} (~{} - ~{})'.format(clazz, int(np.median(low)), int(np.percentile(low, 5)), int(np.percentile(low, 95))))
    print('  {}.2: around {} (~{} - ~{})\n'.format(clazz, int(np.median(high)), int(np.percentile(high, 5)), int(np.percentile(high, 95))))

clazz=3
print('{} class:'.format(clazz))
t = filtered[clazz]
threshold = (t.min() + t.max()) // 2
low_med, high = t[t < threshold], t[t >= threshold]
threshold = (low_med.min() + low_med.max()) // 2
low, med = low_med[low_med < threshold], low_med[low_med >= threshold]
print(' Three groups:')
print('  {}.1: around {} (~{} - ~{})'.format(clazz, int(np.median(low)), int(np.percentile(low, 5)), int(np.percentile(low, 95))))
print('  {}.2: around {} (~{} - ~{})'.format(clazz, int(np.median(med)), int(np.percentile(med, 5)), int(np.percentile(med, 95))))
print('  {}.3: around {} (~{} - ~{})'.format(clazz, int(np.median(high)), int(np.percentile(high, 5)), int(np.percentile(high, 95))))
```

```python
print('Survivors analysis.\n')

is_survived = array(df['Survived'] == 1)
print('Survived: {}% ({}/{} passangers)\n'.format(100 * is_survived.sum() // len(is_survived), is_survived.sum(), len(is_survived)))

classes = df['Pclass']
print('Survivivors percent per class:')
for clazz in np.unique(classes):
    clazz_size = (classes == clazz).sum()
    clazz_survived = (np.logical_and(classes == clazz, is_survived)).sum()
    print(' {} class: {}% ({}/{} passangers)'.format(clazz, 100 * clazz_survived//clazz_size, clazz_survived, clazz_size))
print('So it looks like, that in 3rd class people have had bad path to rescue. \
May be because their berths where located deep in the ship, so their path \
to rescue-boats was the longest, and laid through many other people with 3rd class tickets.\n')
    
plt.figure(figsize=(18,12), dpi=1000)

plt.subplot2grid((2, 2), (0, 0))
plot_survivors_per_class()

is_male = df['Sex'] == 'male'
male_survived, male_not_survived = np.logical_and(is_male, is_survived).sum(), np.logical_and(is_male, ~is_survived).sum()
female_survived, female_not_survived = np.logical_and(~is_male, is_survived).sum(), np.logical_and(~is_male, ~is_survived).sum()
plt.subplot2grid((2, 2), (0, 1));

print('Survivivors percent per sex:')
print(' male:   {}% ({}/{} passangers)'.format(100*male_survived//is_male.sum(), male_survived, male_not_survived.sum()))
print(' female: {}% ({}/{} passangers)'.format(100*female_survived//(~is_male).sum(), female_survived, (~is_male).sum()))
print('So it looks like, that women have had realy high priority in rescue process.\n')

print('Possible conclusion from two previous facts, that in first and second class nearly all females should be rescued. \
And this is really true - two bottom bars approves that.')
plt.bar(array([0, 1]) - .4, [male_survived, female_survived], label='survived', color='green', alpha=.4)
plt.bar(array([0, 1]) - .4, [male_not_survived, female_not_survived], label='survived',
        bottom=[male_survived, female_survived], color='red', alpha=.4)
plt.xticks([0, 1], ['male', 'female'])
plt.ylabel('Passangers count')
plt.legend(loc='upper right');

for i, mask, title in [(0, is_male, 'Only male passangers:'), (1, ~is_male, 'Only female passangers:')]:
    plt.subplot2grid((2, 2), (1, i));
    unique_classes = np.unique(classes)
    survived_per_class = [np.logical_and(classes == c, np.logical_and(mask, is_survived)).sum() for c in unique_classes]
    plt.bar(unique_classes - .4, survived_per_class, label='survived', color='green', alpha=.4)
    not_survived_per_class = [np.logical_and(classes == c, np.logical_and(mask, ~is_survived)).sum() for c in unique_classes]
    plt.bar(unique_classes - .4, not_survived_per_class, label='not survived', bottom=survived_per_class, color='red', alpha=0.4)
    plt.title(title)
    plt.xticks(unique_classes, ['{} class'.format(c) for c in unique_classes])
    plt.ylabel('Passangers count')
    plt.legend(loc='upper left')
```
